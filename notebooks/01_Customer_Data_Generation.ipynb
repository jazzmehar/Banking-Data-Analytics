```python
import pandas as pd
import numpy as np
import random
from faker import Faker
from datetime import datetime, timedelta

fake = Faker("en_IN")

random.seed(42)
np.random.seed(42)
```

```python
NUM_CUSTOMERS = 30000
```

```python
locations = {
    "Delhi":"Delhi",
    "Noida":"Uttar Pradesh",
    "Ghaziabad":"Uttar Pradesh",
    "Gurugram":"Haryana",
    "Faridabad":"Haryana"
}
```

```python
occupation_income = {

    "Student":(0,300000),

    "Teacher":(350000,800000),

    "Engineer":(600000,2000000),

    "Doctor":(1000000,4000000),

    "Business":(800000,5000000),

    "Government Employee":(500000,1500000),

    "Lawyer":(700000,2500000),

    "Manager":(800000,3000000)

}
```

```python
customers = []

for i in range(1, NUM_CUSTOMERS + 1):

    city = random.choice(list(locations.keys()))

    state = locations[city]

    occupation = random.choice(list(occupation_income.keys()))

    income = random.randint(
        occupation_income[occupation][0],
        occupation_income[occupation][1]
    )

    gender = random.choice(["Male","Female"])

    status = random.choices(
        ["Active","Inactive"],
        weights=[95,5]
    )[0]

    customers.append({

        "CustomerID": i,

        "FirstName": fake.first_name(),

        "LastName": fake.last_name(),

        "Gender": gender,

        "DOB": fake.date_of_birth(
            minimum_age=18,
            maximum_age=70
        ),

        "Phone": fake.phone_number(),

        "Email": fake.email(),

        "City": city,

        "State": state,

        "Occupation": occupation,

        "AnnualIncome": income,

        "JoinDate": fake.date_between(
            start_date="-10y",
            end_date="today"
        ),

        "CustomerStatus": status

    })
```

```python
customers_df = pd.DataFrame(customers)

customers_df.head()
```

```python
def random_datetime(start_date):

    end = datetime.now()

    return start_date + timedelta(

        seconds=random.randint(
            0,
            int((end - start_date).total_seconds())
        )

    )
```

```python
def generate_amount(txn_type):

    if txn_type == "UPI Payment":
        return round(random.uniform(20,10000),2)

    elif txn_type == "ATM Withdrawal":
        return random.choice([
            500,
            1000,
            2000,
            5000,
            10000
        ])

    elif txn_type == "Cash Deposit":
        return round(random.uniform(500,50000),2)

    elif txn_type == "Cash Withdrawal":
        return round(random.uniform(500,30000),2)

    elif txn_type == "POS Purchase":
        return round(random.uniform(100,25000),2)

    elif txn_type == "NEFT":
        return round(random.uniform(1000,100000),2)

    elif txn_type == "IMPS":
        return round(random.uniform(500,50000),2)

    elif txn_type == "RTGS":
        return round(random.uniform(200000,1000000),2)

    else:
        return round(random.uniform(20000,150000),2)
```

```python
# Convert TransactionDateTime to datetime objects, coercing errors to NaT
transactions_df['TransactionDateTime'] = pd.to_datetime(transactions_df['TransactionDateTime'], errors='coerce')

# Sort by Account and Date to ensure chronological order per customer
transactions_df.sort_values(
    by=["AccountID", "TransactionDateTime"],
    inplace=True,
    na_position='last'
)

# Reset index after sorting
transactions_df.reset_index(drop=True, inplace=True)

print("Transactions sorted successfully by Account and Date.")
display(transactions_df.head())
```

```python
latest_balance = (
    transactions_df.groupby("AccountID")["BalanceAfterTransaction"]
    .last()
    .reset_index()
)

accounts = accounts.merge(
    latest_balance,
    on="AccountID",
    how="left",
    suffixes=("", "_New")
)

# Fix: Use 'OpeningBalance' instead of 'CurrentBalance' as the fallback
accounts["CurrentBalance"] = accounts["BalanceAfterTransaction"].fillna(
    accounts["OpeningBalance"]
)

accounts.drop(
    columns=["BalanceAfterTransaction"],
    inplace=True
)

print("Account balances updated to reflect the latest transaction states.")
display(accounts[["AccountID", "OpeningBalance", "CurrentBalance"]].head())
```
